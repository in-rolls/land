```python
from indicate import transliterate
from tqdm.auto import tqdm
tqdm.pandas()
```

```python
cache = {}

def hin_to_eng(s):
  eng = ''
  try:
    for x in s.split():
      if x not in cache:
        cache[x] = transliterate.hindi2english(x)
      eng = eng + ' ' + cache[x]
  except:
    cache[x] = ''
    pass
  return eng
```

```python
hin_to_eng('मुसमात देवकी ओझा')
```

```python
temp_df = df.head()
```

```python
temp_df
```

```python
temp_df['eng_name'] = temp_df['name'].progress_apply(lambda x: hin_to_eng(x))
```

```python
temp_df
```

```python
df['eng_name'] = df['name'].progress_apply(lambda x: hin_to_eng(x))
```

```python
df.to_csv('hindi_names_religion_translated.csv', index=False)
```
