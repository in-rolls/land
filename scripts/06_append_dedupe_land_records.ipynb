## Append All Land Records + Dedupe Land Records by Account Number

```python
import pandas as pd
import os
```

```python
# iterate over files, dedupe by account number and add to a list

directory = '../input/bihar_land_records_csv/'

final_df = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    lr_file = pd.read_csv(f, low_memory=False)
    lr_file_dedupe = lr_file.drop_duplicates(subset=['account_no'])
    final_df.append(lr_file_dedupe)
```

```python
final_frame = pd.concat(final_df, axis=0, ignore_index=True)
final_frame.shape
```

```python
final_frame.to_csv("landrecords_deduped.csv", index=False)
```
