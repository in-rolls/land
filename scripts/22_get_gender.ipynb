```python
import pandas as pd
from naampy import in_rolls_fn_gender

import warnings
warnings.filterwarnings('ignore')

FP = "../data/hindi_names_religion_translated.csv.gz"
```

```python
df = (pd
      .read_csv(FP, usecols=["name", "eng_name"])
      # remove श् -- honorific like 'sir'
      .assign(name=lambda df: df["name"].str.replace("श्", "").str.strip())
      .assign(firstname=lambda df: df.eng_name.str.split().str[0])
      .dropna(subset=["firstname"])
      .drop_duplicates("firstname")
      .pipe(lambda df: in_rolls_fn_gender(df, "firstname"))
     )
df.info(show_counts=True)
df.head()
```
