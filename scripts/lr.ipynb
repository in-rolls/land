```python
import pandas as pd
import os
```

```python
directory = '/content/drive/MyDrive/caste/bihar_land_records_csv/'

final_df = []
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    lr_file = pd.read_csv(f, low_memory=False)
    #lr_file_dedupe = lr_file.drop_duplicates(subset=['account_no'])
    final_df.append(lr_file)
```

```python
final_frame = pd.concat(final_df, axis=0, ignore_index=True)
final_frame.shape
```
