```python
import pandas as pd
from naampy import in_rolls_fn_gender

import warnings
warnings.filterwarnings('ignore')

# FP = "../data/hindi_names_religion_translated.csv.gz"
FP = "temp_hindi_names_religion_translated.parquet"
```

```python
df = pd.read_parquet(FP, columns=["name", "eng_name"])
df.info(show_counts=True)
df.head()
```

```python
df.query("eng_name!=eng_name")
```

```python
df.query("eng_name==''")
```

```python
df = (
    df
    .query("eng_name!=''")
    .assign(firstname=lambda df: df.eng_name.str.split().str[0])
)
df.info(show_counts=True)
df.head()
```

```python
df = (
    df
    .drop_duplicates("firstname")
    .pipe(lambda df_: in_rolls_fn_gender(df_, "firstname"))
)
df.info(show_counts=True)
df.head()
```

```python
df.query("pred_gender!=pred_gender")
```

```python
# fill in pred_gender (for those in rolls)
mask = df["pred_gender"].isna()

df.loc[mask & (df["prop_female"] > 0.5), "pred_gender"] = "female"
df.loc[mask & (df["prop_female"] <= 0.5), "pred_gender"] = "male"
df
```
