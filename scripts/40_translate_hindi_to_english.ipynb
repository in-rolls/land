```python
import pandas as pd
from indicate import transliterate
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings('ignore')


INPUT_FILE = '../data/hindi_names_religion.csv.gz'
OUTPUT_FILE = '../data/hindi_names_religion_translated.csv.gz'

df = pd.read_csv(INPUT_FILE, compression='gzip')
df
```

```python
cache = {}

def hin_to_eng(name):
    """Transliterate Hindi name to English with word-level caching."""
    if pd.isna(name):
        return ''
    
    words = []
    for word in str(name).split():
        if word not in cache:
            try:
                cache[word] = transliterate.hindi2english(word)
            except:
                cache[word] = ''
        words.append(cache[word])
    
    return ' '.join(words).strip()
```

```python
tqdm.pandas()
df['eng_name'] = df['name'].progress_apply(hin_to_eng)
```

```python
df.to_parquet("temp_hindi_names_religion_translated.parquet", index=False)
```
