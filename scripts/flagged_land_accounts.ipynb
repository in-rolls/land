```python
import pandas as pd
pd.set_option('display.max_columns', 50)
import janitor
from utilities.utils import get_fulldata
# from indicate import transliterate
pd.set_option('display.max_rows', 100)

# usecols = ['name_of_ryot', 'residence', 'district', 'division', 'caste', 'mouza', 'account_no', '6', '7', '8']
usecols = ['account_no', '6', '7', '8']
usecols = ['name_of_ryot', "name_of_father", "account_no", '6', '7', '8']
df = (
    get_fulldata(**{'usecols':usecols})
#     get_fulldata()
    .rename_columns(new_column_names={"6": "acres", "7": "decimals", "8": "hectare",})
    .assign(tt_area_acre=lambda df: df["acres"]+df["decimals"]/100)
)
df.shape
```

### Large lands

```python
with open("../data/flagged_names.txt") as file:
    # Read the lines of the file and store them in a list
    lines = file.readlines()
names_todrop = [line.strip() for line in lines]
print(f"{len(names_todrop)=}")
names_todrop
```

```python
# top 10,000 land area
_df = (
    df
    .query("name_of_ryot==name_of_ryot")
#     .query("~name_of_ryot.str.contains('|'.join(@names_todrop))")
    .query("name_of_ryot not in @names_todrop")
    .sort_values("tt_area_acre", ascending=False)
    .head(10_000)
    .reset_index(drop=True)
)
_df.to_excel("../10_000_largest_plots.xlsx", index=False)
_df.head(10)
```
