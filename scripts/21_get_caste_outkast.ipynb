```python
import pandas as pd
from outkast import secc_caste

FP = "../data/hindi_names_religion_translated.csv.gz"
```

```python
df = (pd.read_csv(FP, usecols=["name", "eng_name"])
      .assign(lastname=lambda df: [name.split(' ')[-1] for name in df.eng_name])
      .pipe(lambda df: secc_caste(df, "lastname"))
     )
df.head()
```

```python
df.to_csv('../data/hindi_eng_names_caste_outkast.csv.gz', index=False, compression="gzip")
```
